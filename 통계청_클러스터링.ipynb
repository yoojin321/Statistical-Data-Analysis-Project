```python
area = pd.read_csv('area.csv')
sns_mentioned=pd.read_csv('sns_mentioned.csv')
ranking=pd.read_csv('ranking.csv')
```

```python
area.drop(columns='index', inplace=True)
```

```python
area
```

```python
area['지역명'] = area['시도명'] + ' ' + area['시군구명']
```

```python
area = area.groupby('지역명').size().reset_index(name='관광지 수')
```

```python
area
```

```python
sns_mentioned
```

```python
sns_mentioned.drop(columns='Unnamed: 0', inplace=True)
```

```python
sns_mentioned
```

```python
ranking
```

```python
ranking['지역명'] = ranking['광역시/도'] + ' ' + ranking['시/군/구']
```

```python
# '지역명'을 기준으로 그룹화하여 '순위'의 평균 계산
ranking = ranking.groupby('지역명')['순위'].mean().reset_index()

ranking['평균 검색 순위'] = ranking['순위'].round().astype(int)
ranking.drop(columns='순위',inplace=True)
```

```python
ranking
```

```python
visit=pd.read_csv('summed_visitors.csv')
```

```python
visit
```

```python
visit['지역명'] = visit['지자체']
```

```python
visit.drop(columns='지자체',inplace=True)
```

```python
visit
```

```python
merged_df = area.merge(visit, on='지역명')\
                .merge(sns_mentioned, on='지역명')\
                .merge(ranking, on='지역명')
```

```python
merged_df
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
X = merged_df[['관광지 수', '방문자수', '검색량(건)', '평균 검색 순위']]
# 방문자 수에 가중치 부여 (예: 2배)
X['방문자수'] = X['방문자수'] * 2
# 데이터 표준화
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# K-평균 클러스터링 (k=2)
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X_scaled)
```

```python
# 클러스터 라벨을 데이터프레임에 추가
merged_df['클러스터'] = kmeans.labels_

# 각 클러스터의 중심값 계산
cluster_centers = kmeans.cluster_centers_
```

```python
# 각 클러스터의 특성 분석
cluster_summary = pd.DataFrame(cluster_centers, columns=['관광지 수', '방문자수', '검색량(건)', '평균 순위'])
```

```python
cluster_summary
```

```python
# 주류 지역과 비주류 지역 라벨링 (기준: 방문자수)
# 평균 순위가 높을수록 낮은 값이 유리하기 때문에 이 경우 역순으로 정렬
if cluster_summary.loc[0].mean() > cluster_summary.loc[1].mean():
    merged_df['지역유형'] = merged_df['클러스터'].map({0: '주류 지역', 1: '비주류 지역'})
else:
    merged_df['지역유형'] = merged_df['클러스터'].map({0: '비주류 지역', 1: '주류 지역'})
```

```python
merged_df[['지역명', '지역유형']]
```

```python
major = merged_df[merged_df['지역유형'] == '주류 지역']['지역명'].tolist()
minor = merged_df[merged_df['지역유형'] == '비주류 지역']['지역명'].tolist()
```
